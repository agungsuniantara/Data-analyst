```python
# 2. Visualisasi distribusi jumlah ship mode dengan bar plot

store["Ship Mode"].value_counts().plot(kind="bar")
plt.title("Distribusi Jumlah Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Jumlah")
plt.show()
```

```python
# 3. Visualisasikan jumlah penjualan pada masing-masing region dengan bar plot

region_quantity_df = store[["Region", "Quantity"]].copy()
region_quantity_df
```

```python
region_quantity_df.groupby(["Region"]).sum().sort_values(by="Quantity", ascending=False).plot(kind="bar")
plt.title("Jumlah Penjualan per Wilayah")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
# 4. Visualisasikan jumlah profit dan penjualan pada masing2 kategori.

sales_profit_category_df = store[["Category", "Sales", "Profit"]]
sales_profit_category_df
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = "center")
```

```python
profit_by_category = sales_profit_category_df.groupby("Category").sum()["Profit"]
categories = list(profit_by_category.index.values)
profits = list(profit_by_category.values)

profit_by_category
```

```python
plt.figure(figsize=(8, 6))
plt.bar(categories, profits)
addlabels(categories, profits)
plt.xlabel("Category")
plt.ylabel("Sum of Profit")
plt.title("Amount of Profit provided in each category")
plt.show()
```

```python
sales_by_category = sales_profit_category_df.groupby("Category").sum().sort_values("Sales", ascending=False)["Sales"]
categories = list(sales_by_category.index.values)
sales = list(sales_by_category.values)

sales_by_category
```

```python
plt.figure(figsize=(8, 6))
plt.bar(categories, sales)
addlabels(categories, sales)
plt.xlabel("Category")
plt.ylabel("Sum of Sales")
plt.title("Amount of Sales provided in each category")
plt.show()
```

```python
# 5. Visualisasikan hubungan antara discount dengan profit menggunakan scatter plot

discount_profit_df = store[["Discount", "Profit"]]
discount_profit_df
```

```python
plt.scatter(discount_profit_df["Discount"], discount_profit_df["Profit"])
```

```python
plt.plot(discount_profit_df.groupby("Discount").sum())
plt.scatter(discount_profit_df["Discount"], discount_profit_df["Profit"])
plt.show()
```

```python
sum_of_profit_by_discount = discount_profit_df.groupby("Discount").sum()
x = sum_of_profit_by_discount.index.values
y = sum_of_profit_by_discount["Profit"].values

plt.scatter(x,y)
plt.plot(sum_of_profit_by_discount)
plt.show()
```
