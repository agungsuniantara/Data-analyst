```python
import pandas as pd
import matplotlib.pyplot as plt

url = "C:/Users/suniantara/Desktop/bootcamp/data/databaru (1).csv"
data = pd.read_csv(url)
data
```

```python
for col in data.columns:
    print(col)
```

```python
data = data.drop(data.columns[9], axis = 1)
data
```

```python
data[data.columns[7]] = data[data.columns[7]].replace('[\$ ,]', '', regex=True).astype(float)
data
```

```python
data = data.dropna()
data
```

```python
data.dtypes
```

```python
data.hist(column='Price', bins = 267)
```

```python
data.hist(column='Price', by='Type of property', edgecolor='black', range=(100000, 300000))
```

```python
apartment = data[data['Type of property'] == 'Apartment']
apartment
```

```python
mean_apartment = apartment['Price'].mean()
mean_apartment
```

```python
median_apartment = apartment['Price'].median()
median_apartment
```

```python
modus_apartment = apartment['Price'].mode().values[0]
modus_apartment
```

```python
modus_apartment = apartment['Price'].mode()

for i in modus_apartment.index:
     if i > 0:
        modus_apartment = 0
        
modus_apartment.values[0]
```

```python
skewness_apartment = apartment['Price'].skew()
skewness_apartment
```

```python
varians_apartment = apartment['Price'].var()
varians_apartment
```

```python
deviasi_apartment = apartment['Price'].std()
deviasi_apartment
```

```python
office = data[data['Type of property'] == 'Office']
office
```

```python
mean_office = office['Price'].mean()
mean_office
```

```python
median_office = office['Price'].median()
median_office
```

```python
modus_office = office['Price'].mode()

for i in modus_office.index:
     if i > 0:
        modus_office = 0

modus_office
```

```python
skewness_office = office['Price'].skew()
skewness_office
```

```python
varians_office = office['Price'].var()
varians_office
```

```python
deviasi_office = office['Price'].std()
deviasi_office
```

```python
hasil = []
kolom = []
produk = ['Apartment', 'Office']
hasil_akhir = pd.DataFrame(hasil, columns = kolom, index = produk)
hasil_akhir
```

```python
mean = [mean_apartment, mean_office]
median = [median_apartment, median_office]
modus = [modus_apartment, modus_office]
skewness = [skewness_apartment, skewness_office]
varians = [varians_apartment, varians_office]
deviasi = [deviasi_apartment, deviasi_office]
hasil_akhir.insert(loc = 0, column = 'Mean', value = mean)
hasil_akhir.insert(loc = 1, column = 'Median', value = median)
hasil_akhir.insert(loc = 2, column = 'Modus', value = modus)
hasil_akhir.insert(loc = 3, column = 'Skewness', value = skewness)
hasil_akhir.insert(loc = 4, column = 'Varians', value = varians)
hasil_akhir.insert(loc = 5, column = 'Standar Deviasi', value = deviasi)
hasil_akhir
```
