```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import os

files = os.listdir("C:/Users/suniantara/Desktop/bootcamp/data/uber/")
files
```

```python
files.remove('uber-raw-data-janjune-15.csv')
files
```

```python
path = "C:/Users/suniantara/Desktop/bootcamp/data/uber/"

final = pd.DataFrame()

for file in files:
    df = pd.read_csv(path + file,encoding='utf-8')
    final = pd.concat([df,final])
```

```python
final.shape
```

```python
final.info()
```

```python
df = final.copy()
df.head()
```

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format="%m/%d/%Y %H:%M:%S")
```

```python
df.dtypes
```

```python
df['weekday'] = df['Date/Time'].dt.day_name()
df['day'] = df['Date/Time'].dt.day
df['minute'] = df['Date/Time'].dt.minute
df['month'] = df['Date/Time'].dt.month
df['hour'] = df['Date/Time'].dt.hour

df.dtypes
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['day'], bins=31,rwidth=8,range=(0.5,31))
plt.xlabel('date of the month')
plt.ylabel('total journeys')
plt.title('journeys by month day')
```

```python
plt.figure(figsize=(10,10))
for i,month in enumerate(df['month'].unique(),1):
    plt.subplot(3,2,i)
    df_out = df[df['month'] == month]
    plt.hist(df_out['day'], rwidth=0.8)
    plt.xlabel('days in month %i' %month)
    plt.ylabel('total rides')
```
