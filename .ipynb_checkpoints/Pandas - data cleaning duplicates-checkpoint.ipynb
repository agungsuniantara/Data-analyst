```python
import pandas as pd

country= pd.Series([
    'France',
    'United Kingdom',
    'United Kingdom',
    'Italy',
    'Germany',
    'Germany',
    'Germany',
], index=[
    'Country 1',
    'Country 2',
    'Country 3',
    'Country 4',
    'Country 5',
    'Country 6',
    'Country 7'
])

country
```

```python
country.duplicated()
```

```python
country.duplicated(keep="last")
```

```python
country.duplicated(keep=False)
```

```python
country= pd.Series([
  'France',
  'United Kingdom',
  'United Kingdom',
  'Italy',
  'Germany',
  'Germany',
  'Germany',
], index=[
  'Country 1',
  'Country 2',
  'Country 3',
  'Country 4',
  'Country 5',
  'Country 6',
  'Country 7'
])

country
```

```python
country.loc[country.duplicated()] = " "
country
```

```python
country.drop_duplicates()
```

```python
country.drop_duplicates(keep="last")
```

```python
country.drop_duplicates(keep=False)
```

```python
#assignment

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

url = "C:/Users/suniantara/Desktop/bootcamp/data/netflix_titles-1.csv"

data = pd.read_csv(url)
data
```

```python
data = data.set_index('show_id')
data
```

```python
data.duplicated()
```

```python
for i in data.columns:
    data[i][data[i].duplicated()] = np.nan

data
```

```python
for i in data.columns:
    data[i][data[i].isnull()] = " "

data
```
