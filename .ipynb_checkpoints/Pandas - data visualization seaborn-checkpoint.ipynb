```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url="C:/Users/suniantara/Desktop/bootcamp/data/Cause-of-Death-Indonesia.csv"

data2 = pd.read_csv(url)
data2
```

```python
group = data2.groupby('Year')['Total Deaths'].sum()
group
```

```python
plt.figure(figsize=[15,5])
sns.barplot(x='Year', y='Total Deaths',data=group.reset_index(), ci=0)
```

```python
#assignment 2

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url2="C:/Users/suniantara/Desktop/bootcamp/data/Cause-of-Death-Indonesia.csv"

data3 = pd.read_csv(url2)
data3
```

```python
bencana_sosial = data3[data3['Type'] == 'Bencana Sosial']
bencana_sosial
```

```python
bencana_group = bencana_sosial.groupby('Cause')['Total Deaths'].sum()
bencana_group
```

```python
plt.figure(figsize=[15,5])
sns.barplot(x='Total Deaths', y='Cause', data=bencana_group.reset_index(), ci=0)
```
