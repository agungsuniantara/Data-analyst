```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

url = './data/penumpang transjakarta desember 2021.csv'
df = pd.read_csv(url)

# menghilangkan opsi tampilan batasan data

pd.set_option('display.max_rows', None)

# Menampilkan dataset

df
```

```python
# Membersihkan data dengan menghapus karakter ';;;;' pada penamaan kolom 'jumlah_penumpang'

df.rename(columns=lambda x: x.strip(';'), inplace=True)

# tampilkan data

df
```

```python
# Fungsi untuk cek apakah value yang masuk memiliki tipe data str / string menggunakan isinstance
# jika value tipe data str maka ubah value yang memiliki ';' menjadi ''

def remove_semicolon(x):
    if isinstance(x, str):
        return x.replace(';', '')
    return x

# menjalankan fungsi remove_semicolon pada setiap data menggunakan applymap

df = df.applymap(remove_semicolon)

# tampilkan data

df
```

```python
# Membersihkan data dengan menghapus baris yang kosong

df_cleaned = df.dropna()

df_cleaned
```

```python
# Menghapus duplikat

df_hasil = df_cleaned.drop_duplicates()

df_hasil
```

```python
#data 2

url2 = './data/penumpang transjakarta november 2021.csv'
df2 = pd.read_csv(url2)

df2
```

```python
#gabungkan data desember dan data november menggunakan concat
#ignore_index=True digunakan untuk mengatur ulang indeks setelah penggabungan, sehingga indeks tidak berulang.

merged_df = pd.concat([df_hasil, df2], ignore_index=True)

merged_df
```

```python
#normalisasi

scaler = MinMaxScaler()
merged_df['normalisasi'] = scaler.fit_transform(merged_df[['jumlah_penumpang']])

merged_df
```

```python
# Deteksi dan Penanganan Outliers
# Mendeteksi outliers menggunakan metode IQR (Interquartile Range)

merged_df['jumlah_penumpang'] = pd.to_numeric(merged_df['jumlah_penumpang'])

Q1 = merged_df['jumlah_penumpang'].quantile(0.25)
Q3 = merged_df['jumlah_penumpang'].quantile(0.75)
IQR = Q3 - Q1

outliers_mask = (merged_df['jumlah_penumpang'] < Q1 - 1.5 * IQR) | (merged_df['jumlah_penumpang'] > Q3 + 1.5 * IQR)

# Mengambil baris-baris yang merupakan outliers

merged_df[outliers_mask]
```

```python
# Menghapus outliers

df_hasil_bersih = merged_df[~outliers_mask]

df_hasil_bersih
```
