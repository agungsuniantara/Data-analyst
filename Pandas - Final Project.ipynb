```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = 'C:/Users/suniantara/Desktop/bootcamp/data/dataset_final_project.csv'
data = pd.read_csv(url)
data
```

```python
data['Branch'] = data['Branch'].replace(['A', 'B', 'C'], [1,2,3])
data['City'] = data['City'].replace(['Yangon', 'Naypyitaw', 'Mandalay'], [1,2,3])
data['Customer type'] = data['Customer type'].replace(['Normal', 'Member'], [1,2])
data['Product line'] = data['Product line'].replace(['Electronic accessories', 
                                                     'Fashion accessories','Food and beverages', 
                                                     'Health and beauty', 'Home and lifestyle', 
                                                     'Sports and travel'], [1,2,3,4,5,6])
data['Payment'] = data['Payment'].replace(['Cash', 'Credit card', 'Ewallet'], [1,2,3])
data
```

```python
data['Total'] = data['Unit price'] * data['Quantity']
data['Tax'] = data['Total'] * 0.05
data['Final price'] = data['Total'] - data['Tax']
data
```

```python
data.dropna()
```

```python
jumlah = data.groupby('Product line')['Quantity'].sum()

plt.figure(figsize=[5,5])
jumlah.plot(kind='bar', label = 'Quantity')
plt.grid(False)
plt.legend()
```

```python
normal = data[data['Customer type'] == 1]
member = data[data['Customer type'] == 2]

product1 = normal.groupby('Product line')['Customer type'].count()
product2 = member.groupby('Product line')['Customer type'].count()

plt.figure(figsize=[7,5])
sns.barplot(x='Product line',y='Customer type',data=product1.reset_index(),ci=0)
plt.grid(False)

plt.figure(figsize=[7,5])
sns.barplot(x='Product line',y='Customer type',data=product2.reset_index(),ci=0)
plt.grid(False)
```

```python
pay = data.groupby('Payment')['Payment'].count()
plt.figure(figsize=[5,5])
pay.plot(kind='bar', label='Invoice ID')
plt.legend()
```

```python
pie = data.groupby('Payment')['Product line'].sum()
pie.plot(kind='pie')
plt.legend(['satu','dua','tiga'])
```

```python
price = data.groupby('Product line')['Final price'].sum()
price.plot(kind='bar')
```
