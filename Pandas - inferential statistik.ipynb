```python
import pandas as pd
import numpy as np
import scipy.stats as st

url = 'C:/Users/suniantara/Desktop/bootcamp/data/Confidence-intervals-tugas.csv'

data = pd.read_csv(url)
data
```

```python
data.columns = data.columns.str.replace('Unnamed: 12','Year')
data = data.drop(data.columns[0], axis = 1)
data
```

```python
data = data.dropna()
data
```

```python
data_2015 = data[data['Year'] == 2015]
pria_2015 = data_2015[data_2015['Gender'] == 'Male']
penjualan_2015 = pria_2015['SalePrice']
penjualan_2015 = penjualan_2015.replace('[\$ ,]', '', regex=True).astype(float)
penjualan_2015
```

```python
data_2016 = data[data['Year'] == 2016]
pria_2016 = data_2016[data_2016['Gender'] == 'Male']
penjualan_2016 = pria_2016['SalePrice']
penjualan_2016 = penjualan_2016.replace('[\$ ,]', '', regex=True).astype(float)
penjualan_2016
```

```python
# create 90% confidence interval
interval_2015 = st.t.interval(alpha=0.90, df=len(penjualan_2015), loc=np.mean(penjualan_2015), scale=st.sem(penjualan_2015))
interval_2015
```

```python
# create 90% confidence interval
interval_2016 = st.t.interval(alpha=0.90, df=len(penjualan_2016), loc=np.mean(penjualan_2016), scale=st.sem(penjualan_2016))
interval_2016
```
