```python
figure, ax1 = plt.subplots(figsize=(10,6))

ax1.plot(data1['year'],data1['January'], label='januari', c = 'red', alpha = 0.5)
ax2 = ax1.twinx()
ax2.plot(data1['year'],data1['March'], label='maret', alpha=0.5)

ax1.set_xlabel("years")
ax1.set_ylabel("birth rate")
ax2.set_ylabel("population change")
ax1.legend()
ax2.legend()
plt.title("Anderson County birth rate and population change");
```

```python
#assignment

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = 'C:/Users/suniantara/Desktop/bootcamp/data/SampleSuperstore.csv'
data1 = pd.read_csv(url)

data1
```

```python
data1.dropna()
```

```python
group = data1.groupby('Ship Mode')['Quantity'].sum()
group
```

```python
group.plot(kind='bar')
```

```python
plt.figure(figsize=[15,5])
sns.barplot(x='Ship Mode', y='Quantity', data=group.reset_index(), ci=0)
```

```python
region = data1.groupby('Region')['Sales'].sum()
region
```

```python
plt.figure(figsize=[15,5])
sns.barplot(x='Sales', y='Region', data=region.reset_index(), ci=0)
```

```python
profit = data1.groupby('Category')['Profit'].sum()
plt.figure(figsize=[15,5])
sns.barplot(x='Category', y='Profit', data=profit.reset_index(), ci=0)
```

```python
sales = data1.groupby('Category')['Sales'].sum()
plt.figure(figsize=[15,5])
sns.barplot(x='Category', y='Sales', data=sales.reset_index(), ci=0)
```

```python
plt.figure(figsize=(12,6))

plt.scatter(data1['Discount'],data1['Profit'], alpha=0.5, s = 50, label = 'Profit')
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.title("Discount dan Profit")
plt.legend()
```

```python
discount_profit_df = data1[["Discount", "Profit"]]
discount_profit_df
```

```python
plt.scatter(discount_profit_df["Discount"], discount_profit_df["Profit"])
```

```python
plt.plot(discount_profit_df.groupby("Discount").sum())
plt.scatter(discount_profit_df["Discount"], discount_profit_df["Profit"])
plt.show()
```

```python
sum_of_profit_by_discount = discount_profit_df.groupby("Discount").sum()
x = sum_of_profit_by_discount.index.values
y = sum_of_profit_by_discount["Profit"].values

plt.scatter(x,y)
plt.plot(sum_of_profit_by_discount)
plt.show()
```
